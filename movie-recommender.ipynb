```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
```

```python
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies.shape
```

```python
credits.shape
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew','popularity','vote_average']]
```

```python
movies
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.head()
```

```python
movies.iloc[0].genres
```

```python
import ast
```

```python
 def convert(obj):
     objj=ast.literal_eval(obj)
     L=[]
     for i in objj:
         L.append(i['name'])
     return L     
```

```python
convert(movies.iloc[0].genres)
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
 def convert3(obj):
     counter=0
     objj=ast.literal_eval(obj)
     L=[]
     for i in objj:
         L.append(i['name'])
         counter+=1
         if(counter==3):
             break
     return L     
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
 def fetch_director (obj):
     objj=ast.literal_eval(obj)
     L=[]
     for i in objj:
         if(i['job']=='Director'):
             L.append(i['name'])
     return L     
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies['overview']=movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
movies.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
data=movies[['popularity']]
movies['popularity'] = scaler.fit_transform(data)
# scaled_data
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
 movies['tags']= movies['overview']+ movies['genres']+ movies['keywords']+ movies['cast']+ movies['crew']
```

```python
movies.head()
```

```python
new_movies=movies[['movie_id','title','tags']]
```

```python
new_movies
```

```python
new_movies['tags']=new_movies['tags'].apply(lambda x:" ".join(x))
```

```python
new_movies['tags']=new_movies['tags'].apply(lambda x:x.lower())
```

```python
new_movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_movies['tags']).toarray()
```

```python
cv.fit_transform(new_movies['tags']).toarray().shape
```

```python
vectors[3]
```

```python
cv.get_feature_names_out()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    string=" ".join(y)
    return string
```

```python
new_movies['tags']=new_movies['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index=new_movies[new_movies['title']==movie].index[0] 
    distances=similarity[movie_index]   
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6] 
    for i in movies_list:
        print(new_movies.iloc[i[0]].title)
        # print(i[0])
```

```python
recommend('Batman Begins')
```
